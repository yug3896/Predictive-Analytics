```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

beer_df = pd.read_csv( 'beer.csv' )
beer_df.info()
```

```python
beer_df.mean()
```

```python
beer_df.var()
```

```python
# Normalization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_beer_df = scaler.fit_transform( beer_df[['calories',
'sodium',
'alcohol',
'cost']] )
X=pd.DataFrame(scaled_beer_df, index=beer_df.name, 
               columns=['calories','sodium','alcohol','cost'])
```

```python
X
```

```python
from sklearn.decomposition import PCA
pca_loadings = pd.DataFrame(PCA().fit(X).components_.T, 
                            index=['calories','sodium','alcohol','cost'], 
                            columns=['V1', 'V2', 'V3', 'V4'])
pca_loadings
```

```python
from sklearn.decomposition import PCA
pca_loadings = pd.DataFrame(PCA().fit(X).components_.T, 
                            index=['calories','sodium','alcohol','cost'], 
                            columns=['V1', 'V2', 'V3', 'V4'])
pca_loadings

# Fit the PCA model and transform X to get the principal components
pca = PCA()
df_plot = pd.DataFrame(pca.fit_transform(X), columns=['PC1', 'PC2', 'PC3', 'PC4'], 
                       index=X.index)
df_plot
```

```python
fig , ax1 = plt.subplots(figsize=(9,7))

ax1.set_xlim(-3.5,3.5)
ax1.set_ylim(-3.5,3.5)

# Plot Principal Components 1 and 2
for i in df_plot.index:
    ax1.annotate(i, (df_plot.PC1.loc[i], df_plot.PC2.loc[i]), ha='center')

# Plot reference lines
ax1.hlines(0,-3.5,3.5, linestyles='dotted', colors='grey')
ax1.vlines(0,-3.5,3.5, linestyles='dotted', colors='grey')

ax1.set_xlabel('First Principal Component')
ax1.set_ylabel('Second Principal Component')
    
# Plot Principal Component loading vectors, using a second y-axis.
ax2 = ax1.twinx().twiny() 

ax2.set_ylim(-1,1)
ax2.set_xlim(-1,1)
ax2.tick_params(axis='y', colors='red')
ax2.set_xlabel('Principal Component loading vectors', color='red')

# Plot labels for vectors. Variable 'a' is a small offset parameter to separate arrow tip and text.
a = 1.07  
for i in pca_loadings[['V1', 'V2']].index:
    ax2.annotate(i, (pca_loadings.V1.loc[i]*a, pca_loadings.V2.loc[i]*a), color='red')

# Plot vectors
ax2.arrow(0,0,pca_loadings.V1[0], pca_loadings.V2[0])
ax2.arrow(0,0,pca_loadings.V1[1], pca_loadings.V2[1])
ax2.arrow(0,0,pca_loadings.V1[2], pca_loadings.V2[2])
ax2.arrow(0,0,pca_loadings.V1[3], pca_loadings.V2[3]);
```

```python
#The PCA function also outputs the variance explained by of each principal component. 
#We can access these values as follows:
pca.explained_variance_
```

```python
#We can also get the proportion of variance explained:
pca.explained_variance_ratio_
```

```python
"""We see that the first principal component explains 44.9% of the variance in the data, 
the next principal component explains 35.8% of the variance, and so forth. 
We can plot the PVE explained by each component as follows:"""

plt.figure(figsize=(7,5))
plt.plot([1,2,3,4], pca.explained_variance_ratio_, '-o')
plt.ylabel('Proportion of Variance Explained')
plt.xlabel('Principal Component')
plt.xlim(0.75,4.25)
plt.ylim(0,1.05)
plt.xticks([1,2,3,4])
```

```python
"""We can also use the function cumsum(), 
which computes the cumulative sum of the elements of a numeric vector, 
to plot the cumulative PVE:"""

plt.figure(figsize=(7,5))
plt.plot([1,2,3,4], np.cumsum(pca.explained_variance_ratio_), '-s')
plt.ylabel('Proportion of Variance Explained')
plt.xlabel('Principal Component')
plt.xlim(0.75,4.25)
plt.ylim(0,1.05)
plt.xticks([1,2,3,4])
```
